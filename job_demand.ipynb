```python
import pandas as pd 
from datasets import load_dataset
import matplotlib.pyplot as plt 
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()


```

```python
df.job_posted_date = pd.to_datetime(df.job_posted_date)
```

```python
df.info() 
```

```python
df_saudi =df[df['job_country'] == 'Saudi Arabia'].copy()
```

```python
df_saudi
```

```python
df_saudi['job_posted_month'] = df_saudi['job_posted_date'].dt.month_name()
```

```python
df_saudi
```

```python
saudi_pivot= df_saudi.pivot_table(index='job_posted_month' , columns='job_title_short' ,  aggfunc= 'size')
```

```python
saudi_pivot
```

```python
saudi_pivot['total'] = saudi_pivot.sum(axis=1)


saudi_pivot = saudi_pivot.sort_values(by='total', ascending=False)

saudi_pivot = saudi_pivot.drop(columns='total')


```

```python
saudi_pivot
```

```python
saudi_pivot.plot(kind='line') 

```

```python
top3 = df_saudi['job_title_short'].value_counts().head(3)
```

```python
top3
```

```python
top3 = top3.index.to_list()
```

```python
saudi_pivot[top3]
```

```python
saudi_pivot[top3].plot(kind='line') 
plt.title("monthly job posting for data jobs")
plt.xlabel(2023)
plt.ylabel('job count') 
plt.show()
```
